```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv (r'C:\Users\Trang\Desktop\Data Practice\Medicaid Data 2019.csv')
data.head(5)
```

```python
df2019 = pd.DataFrame(data, columns= ['State','Year','Quarter','Product Name','Number of Prescriptions'])
df2019.head(30)
```

```python
Medicaid2019clean = df2019.dropna()
Medicaid2019clean.head(30)
```

```python
df = Medicaid2019clean.sort_values(by = ['State','Quarter','Number of Prescriptions'])
df
```

```python
ad2019 = df[df['Product Name'].str.lower().isin([x.lower() for x in {"Sertraline","Fluoxetine","Paroxetine","Citalopram","Escitalopram","Fluvoxamine","Vortioxetine","Vilazodone"}])]
ad2019
```

```python
data2 = pd.read_csv (r'C:\Users\Trang\Desktop\Data Practice\Medicaid Data 2020.csv')
data2.head(10)
```

```python
df2020 = pd.DataFrame(data2, columns = ['State','Year','Quarter','Product Name','Number of Prescriptions'])
df2020
```

```python
data3 = pd.read_csv (r'C:\Users\Trang\Desktop\Data Practice\Medicaid Data 2021.csv')
data3
```

```python
df2021 = pd.DataFrame (data3, columns = ['state','year','quarter','product_name','number_of_prescriptions'])
df2021
```

```python
df2021.rename(columns = {'state':'State', 'year':'Year', 'quarter':'Quarter', 'product_name':'Product Name', 'number_of_prescriptions':'Number of Prescriptions'}, inplace = True)
df2021
```

```python
df2021 = df2021.sort_values(by = ['State','Quarter','Number of Prescriptions'])
df2021
```

```python
ad2021 = df2021[df2021['Product Name'].str.lower().isin([x.lower() for x in {"Sertraline","Fluoxetine","Paroxetine","Citalopram","Escitalopram","Fluvoxamine","Vortioxetine","Vilazodone"}])]
ad2021
```

```python
ad1 = ad2019.append(ad2020, ignore_index=True)
ad = ad1.append(ad2021, ignore_index=True)
ad
```

```python
AD = ad.groupby(['State','Year','Quarter','Product Name']).size()
AD
```

```python
ad = pd.DataFrame(AD, columns = ['Number of Prescriptions'])
ad
```

```python
ad.to_csv (r'C:\Users\Trang\Desktop\Data Practice\Antidepressants.csv')
```
